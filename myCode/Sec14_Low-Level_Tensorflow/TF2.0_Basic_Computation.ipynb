```
a = tf.constant([1,2,3])
b = tf.constant([4,5,6])
```

```
c = tf.tensordot(a,b, axes=[0,0])
print(f"c: {c}")
```

```
a.numpy().dot(b.numpy())
```

```
import numpy as np
A0 = np.random.randn(3,3)
b0 = np.random.randn(3,1)
c0 = A0.dot(b0)
print(f"c0: {c0}")
```

```
A = tf.constant(A0)
b = tf.constant(b0)
c = tf.matmul(A,b)
print(f"c: {c}")
```

```
A = tf.constant([[1,2],[3,4]])
b = tf.constant(1)
c = A + b
print(f"c: {c}")
```
